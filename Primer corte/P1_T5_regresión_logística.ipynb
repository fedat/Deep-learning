#0Importar librerías

```
import numpy as np #Mejora el soporte para vectores y matrices
import pandas as pd #Estructura de datos (Ciencia de datos)
import matplotlib.pyplot as plt #Para graficar
import seaborn as sns #interfaz de alto nivel para dibujar gráficos estadísticos (basada en matplotlib)

##librería para la regresipinn logística
from sklearn.linear_model import LogisticRegression
```

##0.1Cargar el dataset desde el repo de github

```
df = pd.DataFrame() ##creamos la instancia del dataframe
df = pd.read_csv("https://raw.githubusercontent.com/luisFernandoCastellanosG/Machine_learning/master/Databaset's/horas_estudio_tutorias_vs_aprobacion.csv")
df.head()
```

#1Conocer los datos
##1.2básicos, nulos y dimensiones



```
df.info() ##info basica de la data
df.describe() ##muestra media, minimo, porcentajes
df.isnull().sum ##¿las base de datos cuenta con registros nulos?
df.head()
```

#2Graficando los datos
##2.1 grafico distribución

```
##sns=acrónimo seaborn; 
##sns= acronimo matplot
sns.set(rc={'figure.figsize':(11.7,8.27)}) ##cambiar tamaño del gráfico
sns.distplot(df['aprobo_perdio'])  ##mostrar datos que deseo fabricar
plt.show()
```

##2.1Gráfico en x y y

```
x=df['horas_autoestudio_diario']
y=df['aprobo_perdio']
##scatter=gráficas en 2D
plt.scatter(x,y,s=15, alpha=0.5)
plt.xlabel('horas estudio')
plt.ylabel('aprobo o reprobo')
plt.show()
```

#3Separar los datos (trains/test)

```
##obtener los datos limpios en x y los colocamos en una sola dimensión
##se limpian los datos para que no generen conflicto
xtrain=np.array(df['horas_autoestudio_diario']).reshape(-1,1)
ytrain=np.array(df['aprobo_perdio']).reshape(-1,1)

print(ytrain)
```

##4Crear instancia del modelo y entrenar

```
reg_logistica=LogisticRegression()
reg_logistica.fit(xtrain,ytrain)
```

#5Haciendo predicciones

```
 x_horas_estudio=np.array([1,2,4,10,25,30]). reshape(-1,1)
 prediccion_ap=reg_logistica.predict(x_horas_estudio)
 print(prediccion_ap)
```

##5.1Cuáles son las probabilidades o inferencias

```
probab_prediccion=reg_logistica.predict_proba(x_horas_estudio)
print(probab_prediccion[:,1])
```
